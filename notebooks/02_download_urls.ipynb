```python
from pathlib import Path
import pandas as pd
from src.gdelt_pipeline.config import load_config
from src.gdelt_pipeline.io import read_csv, save_csv, dedupe_urls
from src.gdelt_pipeline.url_clean import normalize_url, add_domain

settings, paths = load_config()
art = read_csv(paths.interim_dir / "artlist_all.csv")
art.head()
```

```python
if "url" not in art.columns:
    raise ValueError("Expected a 'url' column in artlist_all.csv")

art["url"] = art["url"].astype(str).apply(normalize_url)
art = add_domain(art, url_col="url")
art = dedupe_urls(art, url_col="url")
save_csv(art, paths.interim_dir / "artlist_all_deduped.csv")
len(art)
```
